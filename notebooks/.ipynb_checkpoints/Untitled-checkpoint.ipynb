```python
import pandas as pd
import math
import collections
```

```python
def df2table(df):
    for label in labels:
        print(label)
        lst = []
        for idx, row in df.iterrows():
            ctr = collections.Counter(row)
            if label == None:
                is_nan = [0 for lab in row if math.isnan(lab)]
                num_label = len(is_nan)
            else:
                num_label = ctr[label]
            lst.append(num_label)
        df_tmp[label] = lst
        print(lst)
    return pd.DataFrame(df_tmp)
```
